```python
import networkx as nx
import pylab
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
directory_name = './data/'
input_files = [
    'ksiezyc.csv',
    'noc.csv',
    'pelnia.csv',
    'slonce.csv'
]

edge_list = []

for input_file in input_files:
    with open(directory_name + input_file, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        parsed_list = list(reader)
        for item in parsed_list:
            edge_name = input_file.replace('.csv', '')
            edge_list.append((edge_name, item[1], int(item[0])))
```

```python
edge_list
```

```python
 G = nx.Graph()
```

```python
G
```

```python
G.add_weighted_edges_from(edge_list)
```

```python
G
```

```python
G.edges
```

```python
plt.figure(figsize=(30,30))
options = {'edge_color': '#AAAAAA', 'width': 1, 'with_labels': True}

sizes = [sum(c[2]['weight'] for c in G.edges(node, data=True)) for node in G]

positions = nx.spring_layout(G, k=0.3, iterations=300)
nx.draw(G, pos=positions, node_size=sizes, **options)
```
